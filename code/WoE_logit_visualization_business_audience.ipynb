```python
import random
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Define Tableau 10 Colors
tableau_colors = [
    (31, 119, 180),  # Blue
    (255, 127, 14),  # Orange
    (44, 160, 44),   # Green
    (214, 39, 40),   # Red
    (148, 103, 189), # Purple
    (140, 86, 75),   # Brown
    (227, 119, 194), # Pink
    (127, 127, 127), # Gray
    (188, 189, 34),  # Yellow
    (23, 190, 207),  # Cyan
]

# Normalize RGB values to range [0, 1]
tableau_colors = [(r / 255, g / 255, b / 255) for r, g, b in tableau_colors]
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-(10*(x-0.5))))

def sigmoid2(x):
    return 1 / (1 + np.exp(-(20*(x-0.5))))

def sigmoid3(x):
    return 1 / (1 + np.exp(-(6*(x-0.5))))

x = np.linspace(0, 1, 400)
y1 = sigmoid(x)
y2 = sigmoid2(x)
y3 = sigmoid3(x)

plt.figure(figsize=(8, 8))
ax = plt.gca()
plt.plot(x, y1, linewidth=3, color=tableau_colors[0])
plt.plot(x, y2, linewidth=3, color=tableau_colors[1])
plt.plot(x, y3, linewidth=3, color=tableau_colors[2])

font = {'family': 'Georgia', 'color':  'black', 'weight': 'normal', 'size': 20}

plt.title('Family of Sigmoid Functions', fontdict=font)
plt.xlabel('Normalized Predictive Attribute', fontdict=font)
plt.ylabel('Probability', fontdict=font)
# plt.legend()
# plt.grid(True)

# Customize tick sizes
ax.tick_params(axis='both', which='major', labelsize=18, length=6)  # Adjust label size and tick length

plt.savefig('sigmoid_family.png') 
plt.show()
```

```python
x = np.linspace(0, 1, 400)
y = sigmoid(x)
y2 = sigmoid2(x)
y3 = sigmoid3(x)

plt.figure(figsize= (8, 8))
ax = plt.gca()
plt.plot(x, y, linewidth=4, color=tableau_colors[2])
plt.plot(x, y2, linewidth=2, color=tableau_colors[7], linestyle='-.')
plt.plot(x, y3, linewidth=2, color=tableau_colors[7], linestyle='--')
plt.scatter(xrands0, yrands0, s=120, edgecolors=tableau_colors[9], facecolors='none', marker='o')
plt.scatter(xrands1, yrands1, s=120, edgecolors=tableau_colors[3], facecolors='none', marker='o')

font = {'family': 'Georgia', 'color':  'black', 'weight': 'normal', 'size': 20}

plt.xlabel('Normalized Predictive Attribute', fontdict=font)
plt.ylabel('Probability', fontdict=font)
plt.title('Logistic Regression on Data', fontdict=font)
# plt.legend()
# plt.grid(True)

# Customize tick sizes
ax.tick_params(axis='both', which='major', labelsize=18, length=6)  # Adjust label size and tick length

plt.savefig('data_sigmoid_combined.png') 
plt.show()
```

```python
# calculate original WoE, log-odds
data_dict_0 = Counter(xrands0)
data_dict_1 = Counter(xrands1)

lnodds = []
WoEs   = []

for i in range(101):
    num = i/100
    if (num in data_dict_0) and (num in data_dict_1):
#         odds is good / bad
        ratio = data_dict_0[num] / data_dict_1[num]
        lnodd = np.log(ratio)
        WoE   = np.log((data_dict_0[num] / len(xrands0)) / (data_dict_1[num] / len(xrands1)))
        lnodds.append(lnodd)
        WoEs.append(WoE)
    elif (num in data_dict_0) and (num not in data_dict_1):
        lnodd = 8
        WoE   = 8
        lnodds.append(lnodd)
        WoEs.append(WoE)
    elif (num not in data_dict_0) and (num in data_dict_1):
        lnodd = -8
        WoE   = -8
        lnodds.append(lnodd)
        WoEs.append(WoE)
    else:
        lnodds.append(np.nan)
        WoEs.append(np.nan)
        
# print(lnodds, WoEs)
```
